```python
import pandas as pd
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
carspec_df = pd.read_csv("./ridesharing/carspec2010.csv", header = 2)
```

```python
carspec_df['Brand'] = carspec_df['Brand'].ffill(axis = 0)
carspec_df.head(200)
```

```python
carspec_df = carspec_df[['Brand', 'Model series']]
carspec_df.head(200)
```

```python
# type(carspec_df['Model series'][10])
def removeAfterParenthesis(element):
    if type(element) == str:
        if('(' not in element): 
            return element
        first, second = element.split('(', 1)
        print(first, second)
        return first
    
carspec_df['Model series'] = carspec_df['Model series'].apply(removeAfterParenthesis)
```

```python
carspec_df.drop_duplicates(subset =['Brand', 'Model series'], keep='first', inplace=True)
```

```python
carspec_df.to_csv('./Ridesharing/Output/CarSpec_Brands_Models.csv', sep=',', index=False,header=True)
```
